```python
x = np.arange(10,20)
y = np.arange(100, 110)
```

```python
plt.plot(x,y,color="#6A3B3B", linestyle="dotted", marker="*", ms=10, mec="#023920", mfc="#00ff00")
# plt.plot(x,y)
plt.title("Plot Title")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid()
# plt.grid(axis="x")
# plt.grid(axis="y")
plt.show()
```
