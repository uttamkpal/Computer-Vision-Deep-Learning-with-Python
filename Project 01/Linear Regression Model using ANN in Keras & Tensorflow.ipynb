### Linear Regression Model using ANN in Keras & Tensorflow

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
X = np.arange(1,50,5)
```

```python
print(X)
print(len(X))
```

```python
Y = X + 10
print(Y)
print(len(Y))
```

```python
plt.scatter(X,Y)
```

```python
linear_regression_model = tf.keras.Sequential()
```

```python
linear_regression_model.add(tf.keras.layers.Dense(10, input_shape=[1], name='input_layer'))
linear_regression_model.add(tf.keras.layers.Dense(1, name='output_layer'))
```

```python
linear_regression_model.compile(loss='mae', optimizer='sgd', metrics=['mae'])
```

```python
linear_regression_model.fit(X, Y, epochs=10)
```

```python
linear_regression_model.predict(np.array([9]))
```

### Large amount of data use in the ANN

```python
x = np.arange(1,1000,5)
y = x + 25
print(x, y)
```

```python
lrmld = tf.keras.Sequential()

lrmld.add(tf.keras.layers.Dense(100, input_shape=[1], activation='relu', name='input_layer'))
lrmld.add(tf.keras.layers.Dense(1, activation='linear', name='output_layer'))

adam = tf.keras.optimizers.Adam(learning_rate=0.01)
lrmld.compile(loss='mae', optimizer=adam, metrics=['mae'])

lrmld.fit(x, y, epochs=300)

lrmld.predict(np.array([9]))
```

```python
lrmld = tf.keras.Sequential()

lrmld.add(tf.keras.layers.Dense(100, input_shape=[1], activation='relu'))
lrmld.add(tf.keras.layers.Dense(50, input_shape=[1], activation='relu'))
lrmld.add(tf.keras.layers.Dense(1, activation='linear'))

adam = tf.keras.optimizers.Adam(learning_rate=0.01)
lrmld.compile(loss='mae', optimizer=adam, metrics=['mae'])

lrmld.fit(x, y, epochs=300)

lrmld.predict(np.array([9]))
```
